```python
import pandas as pd
import numpy as np
from collections import Counter

def compare_excel_columns(file1, file2, column1, column2, quantity_col1, quantity_col2, rate_col1, rate_col2, output_file, unused_rows_file):
    # Read the Excel files
    df1 = pd.read_excel(file1)
    df2 = pd.read_excel(file2)

    # Check for duplicates in ODN NO column (file1)
    odn_no_counts = Counter(df1[column1])
    duplicates_file1 = {odn: count for odn, count in odn_no_counts.items() if count > 1}

    if duplicates_file1:
        print("Duplicate ODN NO values found in file1:")
        for odn, count in duplicates_file1.items():
            print(f"ODN NO: {odn}, Repeated: {count} times")
    else:
        print("No duplicate ODN NO values found in file1.")

    # Check for duplicates in Inv No column (file2)
    inv_no_counts = Counter(df2[column2])
    duplicates_file2 = {inv: count for inv, count in inv_no_counts.items() if count > 1}

    if duplicates_file2:
        print("\nDuplicate Inv No values found in file2:")
        for inv, count in duplicates_file2.items():
            print(f"Inv No: {inv}, Repeated: {count} times")
    else:
        print("\nNo duplicate Inv No values found in file2.")

    # Create a dictionary for faster searching
    search_dict = df2.set_index(column2).to_dict('index')

    # Create a set to keep track of used rows in df2
    used_rows = set()

    # Initialize lists to store results
    matched = []
    unmatched = []

    # Create new columns for match results
    df1['Quantity Match'] = ''
    df1['Rate Match'] = ''
    df1['Match Result'] = ''

    # Perform the comparison
    for idx, row in df1.iterrows():
        value = row[column1]
        if value in search_dict:
            used_rows.add(value)
            df2_row = search_dict[value]
            quantity_diff = abs(row[quantity_col1] - df2_row[quantity_col2])
            rate_diff = abs(row[rate_col1] - df2_row[rate_col2])

            # Check quantity match
            if quantity_diff <= 0.2:
                quantity_match = "Exact" if quantity_diff == 0 else f"Close (diff: {quantity_diff:.2f})"
                df1.at[idx, 'Quantity Match'] = f"Match - {quantity_match}"
            else:
                df1.at[idx, 'Quantity Match'] = f"Mismatch (diff: {quantity_diff:.2f})"

            # Check rate match
            if rate_diff == 0:
                rate_match = "Exact"
                df1.at[idx, 'Rate Match'] = f"Match - {rate_match}"
            elif 1 <= rate_diff <= 10:
                rate_match = f"Close (diff: {rate_diff:.2f})"
                df1.at[idx, 'Rate Match'] = f"Match - {rate_match}"
            else:
                df1.at[idx, 'Rate Match'] = f"Mismatch (diff: {rate_diff:.2f})"

            # Determine overall match result
            if quantity_diff <= 0.2 and rate_diff <= 10:
                df1.at[idx, 'Match Result'] = "Full Match"
                matched.append(f"Row {idx + 1} in {file1} - Full match for value '{value}' - Quantity: {quantity_match}, Rate: {rate_match}")
            else:
                df1.at[idx, 'Match Result'] = "Partial Match"
                unmatched.append({
                    'row_number': idx + 1,
                    'unmatched_value': value,
                    'quantity_diff': quantity_diff,
                    'rate_diff': rate_diff,
                    'row_details': row.to_dict()
                })
        else:
            df1.at[idx, 'Quantity Match'] = "No match found"
            df1.at[idx, 'Rate Match'] = "No match found"
            df1.at[idx, 'Match Result'] = "No match found"
            unmatched.append({
                'row_number': idx + 1,
                'unmatched_value': value,
                'row_details': row.to_dict()
            })

    # Save the results to a new Excel file
    df1.to_excel(output_file, index=False)

    # Find unused rows in df2
    unused_df = df2[~df2[column2].isin(used_rows)]

    # Save unused rows to a separate CSV file
    unused_df.to_csv(unused_rows_file, index=False)

    # Print matched results
    print("\nMatched Results:")
    for result in matched:
        print(result)

    # Print unmatched results with full row details
    print("\nUnmatched Results:")
    for item in unmatched:
        print(f"Row {item['row_number']} in {file1} - Partial or no match for value '{item['unmatched_value']}'")
        if 'quantity_diff' in item:
            print(f"Quantity difference: {item['quantity_diff']:.2f}")
        if 'rate_diff' in item:
            print(f"Rate difference: {item['rate_diff']:.2f}")
        print("Row Details:")
        for key, value in item['row_details'].items():
            print(f"  {key}: {value}")
        print()

    # Print summary
    print(f"Total Rows in {file1}: {len(df1)}")
    print(f"Fully Matched Rows: {len(matched)}")
    print(f"Partially Matched or Unmatched Rows: {len(unmatched)}")
    print(f"Unused Rows in {file2}: {len(unused_df)}")
    print(f"Results saved to {output_file}")
    print(f"Unused rows from {file2} saved to {unused_rows_file}")

# File paths and column names
file1 = '/content/tar.xlsx'
file2 = '/content/test.xlsx'
column1 = 'ODN NO'  # Column name in the first file
column2 = 'Inv No'  # Column name in the second file
quantity_col1 = 'Quantity'  # Quantity column name in the first file
quantity_col2 = 'LD WT'  # Quantity column name in the second file
rate_col1 = 'Frt/MT'  # Rate column name in the first file
rate_col2 = 'Rate/Ton'  # Rate column name in the second file
output_file = '/content/comparison_results.xlsx'
unused_rows_file = '/content/unused_rows.csv'

compare_excel_columns(file1, file2, column1, column2, quantity_col1, quantity_col2, rate_col1, rate_col2, output_file, unused_rows_file)

# Print the contents of the unused rows file
print("\nContents of the unused rows file:")
unused_df = pd.read_csv(unused_rows_file)
print(unused_df)

```
